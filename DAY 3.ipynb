```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
sm=pd.read_csv('student_marks.csv')
sm
```

```python
sm['Total']=sm[['English','Hindi','Maths','Science','SST']].sum(axis=1)
sm
```

```python
sm['percentage']=sm['Total']/5
sm
```

```python
Grading
A=100-80
B=80-60
C=60-50
D= 50-40
E=40-30
F=0-30
```

```python
def clac_grade(p):
    if p<= 100 and p>=80:
        return 'A'
    elif p<80 and p>=60:
        return 'B'
    elif p<60 and p>=50:
        return 'C'
    elif p<50 and p>=40:
        return 'D'
    if p<= 40 and p>=30:
        return 'E'
    if p<= 30 and p>=0:
        return 'F'
    else:
        return 'X'
    

sm['grade']=sm['percentage'].apply(clac_grade)
sm
```

```python
plt.figure(figsize=(10,6))
for sub in ('English','Hindi','Maths','Science','SST'): 
    plt.plot(sm['Name'],sm[sub],marker='o',label=sub)
    plt.title("Marks in different subjects")
    plt.xlabel('Name')
    plt.ylabel('Marks')
    plt.legend()
    plt.grid(True)
    plt.show()
```
